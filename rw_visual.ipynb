```python
import matplotlib.pyplot as plt
```

```python
from random_walk import RandomWalk
```

```python
# Keep making new walks, as long as the program is active
while True:
    # Make a random walk
    rw = RandomWalk(num_points=50_000) # change number of points
    rw.fill_walk()

    # Plot the points in the walk
    plt.style.use('classic')
    # adjust size of plotting window; dpi = dots per ink, 100 as default
    fig,ax = plt.subplots(figsize=(15,9), dpi=125) 
    point_numbers = range(rw.num_points) # create order of points (type=range)
    ax.scatter(
        rw.x_values, rw.y_values, 
        c=point_numbers, # colormap by order of points, not by xy coordinates
        cmap=plt.cm.Blues,
        edgecolors='none', # remove black outline from each dot
        s=1)
    ax.set_aspect('equal') # -> specify that both axes should have equal spacing between tick marks

    # Emphasize first and last point
    ax.scatter(rw.x_values[0], rw.y_values[0], color='Green', edgecolors='none', s=30)
    ax.scatter(rw.x_values[-1], rw.y_values[-1], color='Red', edgecolors='none', s=30)

    # Make axis visible or not
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    plt.show()

    active_flag = input('Make another walk? (y/n): ')
    if active_flag == 'n':
        break
```
