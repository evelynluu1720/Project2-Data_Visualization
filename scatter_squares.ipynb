```python
import matplotlib.pyplot as plt
```

```python
# Create list manually
x_values_manual = [1,2,3,4,10]
y_values_manual = [1,4,9,16,25]

# Auto create list
x_values = range(1,1001)
y_values = [x**2 for x in x_values]
```

```python
plt.style.use('ggplot')
fig,ax = plt.subplots()
# scatter(x coord, y coord, dot color (RGB 0-1), dot size)
ax.scatter(
    x_values, y_values, 
    # color=(0, 0.8, 0), 
    c=y_values, # specify how pyplot should assign color to each point based on y-value
    cmap=plt.cm.Blues, # use one of colormaps = gradient of colors
    s=10) 

# Set chart title & axes labels
ax.set_title('Square Numbers', fontsize = 20)
ax.set_xlabel('Value', fontsize = 10)
ax.set_ylabel('Square of Value', fontsize = 10)
ax.tick_params(labelsize = 10)

# Set range for each axis
ax.axis([0,1100,0,1_100_000]) # axis(xmin, xmax, ymin, ymax)
ax.ticklabel_format(style='plain') # replace scientific notation with plain numbers

# show plot
# plt.show()

# save plot to file
plt.savefig('squares_plot.png', bbox_inches='tight') # bbox_inches='tight' -> trims whitespace from plot
```
