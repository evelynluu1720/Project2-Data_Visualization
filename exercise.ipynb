```python
import matplotlib.pyplot as plt
```

```python
x_values = [i for i in range(1,5001)]
y_values = [i**3 for i in x_values]
```

```python
fig,ax = plt.subplots()
plt.style.use('ggplot')

ax.scatter(x_values, y_values, c=x_values, cmap=plt.cm.Reds, s=5)
ax.set_title('Cube Numbers', size=20)
ax.set_xlabel('Value', size=10)
ax.set_ylabel('Cube of Value', size=10)
ax.tick_params(labelsize=10)

ax.axis([0,5200,0,140_000_000_000])
ax.ticklabel_format(style='scientific')

plt.show()
```
